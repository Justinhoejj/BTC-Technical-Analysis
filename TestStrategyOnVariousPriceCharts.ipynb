This program tests the performance of each strategy on various price charts over a time period of 1 year. Time period of Feb 2020 - Feb 2021 was chosen to obtain a variety of price charts.

```python
from technicalIndicatorUtils import analyse, execute_MACD, execute_OBV_EMA, execute_MACD_OBV, plot_close_price, execute_buy_hold
import pandas as pd
import numpy as np
```

```python
def report(filepath):
    df = analyse(filepath).head(760).tail(365)
    plot_close_price(df)
    print("\n===== BASELINE BUY AND HOLD STRATEGY ========")
    print(execute_buy_hold(df))
    print("\n===== MACD STRATEGY =========================")
    print(execute_MACD(df))
    print("\n===== MACD STRATEGY WITH STOP LOSS ===========")
    print(execute_MACD(df, use_stop_loss=True))
    print("\n===== OBV STRATEGY ===========================")
    print(execute_OBV_EMA(df))
    print("\n===== OBV STRATEGY WITH STOP LOSS ============")
    print(execute_OBV_EMA(df, use_stop_loss=True))
    print("\n===== OBV & MACD STRATEGY ====================")
    print(execute_MACD_OBV(df))
    print("\n===== OBV & MACD STRATEGY WITH STOP LOSS =====")
    print(execute_MACD_OBV(df, use_stop_loss=True))
```

<h1>Bitcoin (Uptrend with multiple All Time High)</h1>

```python
report('testDataset/BTC-USD.csv')
```

<h3>Observations:</h3>

- Bitcoin price maintained a strong uptrend.
- None of the indicators manage to outperform the buy and hold strategy by percentage gain.
- MACD strategy no significant difference in performance with stop loss.
- OBV EMA strategy is the most aggressive placing over 20 trades majority are losing trades.
- OBV EMA stop loss causes trades to close early deteriorates performance significantly.
- OBV MACD has the best win rate.
- OBV MACD with stop loss is the best performing indicator with compounded gains exceeding buy and hold.
